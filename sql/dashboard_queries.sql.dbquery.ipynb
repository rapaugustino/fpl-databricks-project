```sql
-- ============================================================
-- Dashboard Queries
-- These queries power the basic Databricks SQL dashboard.
-- Run them in the SQL Editor and pin the results to a dashboard.
-- ============================================================

-- Top 20 players by points-per-million value
SELECT display_name, team_name, position, current_price,
       total_points, points_per_million
FROM player_value_rankings
LIMIT 20;

-- Average points-per-million by position (GKP, DEF, MID, FWD)
SELECT position,
       ROUND(AVG(points_per_million), 2) AS avg_ppm,
       COUNT(*) AS player_count
FROM player_value_rankings
GROUP BY position
ORDER BY avg_ppm DESC
```
